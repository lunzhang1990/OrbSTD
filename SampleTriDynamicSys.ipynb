```python
# if index = 1 we have flow from s to t
def genDirectedEdge(s,t,index,d):
    if s not in d:
        d[s] = []
    if t not in d:
        d[t] = []
    
    if index == 1:
        d[s].append(t)
    elif index == 2:
        d[s].append(t)
        d[t].append(s)
    else:
        d[t].append(s)


def Index2STD(l):
    d =  dict()
    # for outer boundary edge
    d[1]=[5]
    d[2]=[7]
    d[3]=[9]
    d[4]=[5]
    d[13]=[14]
    d[22]=[23]
    
    # for edge 1
    genDirectedEdge(6,5,l[0],d)
    genDirectedEdge(10,5,l[0],d)
    genDirectedEdge(7,6,l[1],d)
    genDirectedEdge(14,10,l[1],d)
    
    # for edge 2
    genDirectedEdge(11,7,l[2],d)
    genDirectedEdge(15,14,l[2],d)
    genDirectedEdge(16,11,l[3],d)
    genDirectedEdge(16,15,l[3],d)
    
    # for edge 3
    genDirectedEdge(8,7,l[4],d)
    genDirectedEdge(19,14,l[4],d)
    genDirectedEdge(9,8,l[5],d)
    genDirectedEdge(23,19,l[5],d)
    
    # for edge 4
    genDirectedEdge(17,16,l[6],d)
    genDirectedEdge(20,16,l[6],d)
    genDirectedEdge(18,17,l[7],d)
    genDirectedEdge(25,20,l[7],d)
    
    # for edge 5
    genDirectedEdge(12,9,l[8],d)
    genDirectedEdge(26,25,l[8],d)
    genDirectedEdge(18,12,l[9],d)
    genDirectedEdge(25,24,l[9],d)
    
    # for edge 6
    genDirectedEdge(21,18,l[10],d)
    genDirectedEdge(26,25,l[10],d)
    genDirectedEdge(27,26,l[11],d)
    genDirectedEdge(27,21,l[11],d)
    
    return  d
```
